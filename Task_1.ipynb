```python
# ==============================
# Task 1: Business Sales Performance Analytics
# Google Colab - Single Cell Code
# ==============================

import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# Create Dummy Sales Dataset
# ------------------------------
data = {
    "order_date": [
        "2024-01-05","2024-01-10","2024-01-15","2024-01-20",
        "2024-02-02","2024-02-08","2024-02-14","2024-02-18",
        "2024-03-01","2024-03-06","2024-03-10","2024-03-15",
        "2024-04-02","2024-04-08","2024-04-12","2024-04-18"
    ],
    "region": [
        "South","North","East","West",
        "South","North","East","West",
        "South","North","East","West",
        "South","North","East","West"
    ],
    "product": [
        "Laptop","Mobile","Headphones","Tablet",
        "Monitor","Laptop","Mobile","Headphones",
        "Tablet","Monitor","Laptop","Mobile",
        "Headphones","Tablet","Monitor","Laptop"
    ],
    "category": [
        "Electronics","Electronics","Accessories","Electronics",
        "Electronics","Electronics","Electronics","Accessories",
        "Electronics","Electronics","Electronics","Electronics",
        "Accessories","Electronics","Electronics","Electronics"
    ],
    "quantity": [2,3,5,1,2,1,4,6,2,3,1,2,4,1,2,1],
    "unit_price": [52000,22000,2500,30000,15000,50000,21000,2000,
                   32000,14000,53000,23000,2800,31000,16000,51000]
}

df = pd.DataFrame(data)

# Convert date & calculate revenue
df["order_date"] = pd.to_datetime(df["order_date"])
df["revenue"] = df["quantity"] * df["unit_price"]

print("DATASET PREVIEW")
display(df.head())

# ------------------------------
# Monthly Revenue Trend
# ------------------------------
df["month"] = df["order_date"].dt.to_period("M")
monthly_revenue = df.groupby("month")["revenue"].sum()

monthly_revenue.plot(title="Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

# ------------------------------
# Top Products
# ------------------------------
top_products = df.groupby("product")["revenue"].sum().sort_values(ascending=False)

top_products.plot(kind="bar", title="Top Selling Products")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()

# ------------------------------
# Revenue by Category
# ------------------------------
category_revenue = df.groupby("category")["revenue"].sum()

category_revenue.plot(kind="bar", title="Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()

# ------------------------------
# Revenue by Region
# ------------------------------
region_revenue = df.groupby("region")["revenue"].sum()

region_revenue.plot(kind="bar", title="Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.show()

# ------------------------------
# Key Insights
# ------------------------------
print("KEY INSIGHTS:")
print("1. Electronics category contributes the highest revenue.")
print("2. Laptop and Mobile are top revenue-generating products.")
print("3. South and North regions perform better in sales.")
print("4. Monthly revenue shows stable business performance.")

# ------------------------------
# Save Dataset (for GitHub)
# ------------------------------
df.to_csv("sales_data.csv", index=False)
print("sales_data.csv file created successfully.")
```
